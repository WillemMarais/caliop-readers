# CALIOP L1B reader example

## Boiler plate code

```python
from pathlib import Path

from caliopreaders import CALIOPlevel1B
from caliopreaders.data import create_sim_file_registery
```

## Get sample L1B data

```python
sim_file_registery = create_sim_file_registery()

# Get the sample L1B file
lvl1B_fileP = Path(
    sim_file_registery.fetch(
        "CAL_LID_L1-Standard-V4-10.2016-09-18T13-13-48ZD.hdf"
    )
)
```

## Load the L1B xarray dataset

```python
# Load the L1B and create an xarray dataset
l1b_reader = CALIOPlevel1B(
    lvl1B_fileP_lst=[
        lvl1B_fileP
    ]
)

l1b_ds = l1b_reader.to_xr(
    # Include meteorogical data such as the 
    # molecular and ozone extinctions; this 
    # make the reader slow. To open the 
    # L1B faster, disable this flag
    incld_met_data_bl=True
)

l1b_ds
```
